```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
```

```python
df = pd.read_csv("salary_prediction_kaggle//Salary.csv")
df.head()
df.shape
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot(df["YearsExperience"], df["Salary"])  # Plot some data on the axes.
```

```python
X_data, Y_data = df["YearsExperience"], df["Salary"] 
X_data= X_data.values.reshape(-1,1)
X_train, X_test, Y_train, Y_test = train_test_split(X_data,Y_data.values, train_size=0.8)
```

```python
regr = LinearRegression()
regr.fit(X_train, Y_train.ravel())
```

```python
Y_test_predict = regr.predict(X_test)
print(Y_test_predict)
print(Y_test)
```

```python
fig2, ax2 = plt.subplots()
ax2.scatter(X_test, Y_test,color='b')
ax2.scatter(X_test,Y_test_predict,color= 'r')
```

```python
rmse = MSE(Y_test_predict, Y_test)**(1/2)
print(f"RMSE is {rmse}")
print(f"Median of the data is {Y_data.median()} and error is {rmse/Y_data.median()*100}%")

```
