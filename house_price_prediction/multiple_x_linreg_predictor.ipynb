```python
import pandas as pd
import numpy as np
import sklearn as skl
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
```

```python
ary = np.loadtxt("boston.csv", delimiter=",")
X_values = ary[:,:13]
Y_values = ary[:,13]
X_features= ["CROM", "ZN", "INDUS",'CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
```

```python
fig, axes = plt.subplots(3,5,layout="constrained",figsize = (20,20))
count = 0
for i in range(3):
    for j in range(5):
        axes[i][j].scatter(X_values[:,count],Y_values)
        axes[i][j].set_title(X_features[count])
        count+= 1
        if count == 13:
            break
```

```python
regr = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(X_values,Y_values, test_size=0.2)
x_train.shape
```

```python
regr.fit(x_train,y_train)
y_predict = regr.predict(x_test)
```

```python
fig,axes = plt.subplots()
axes.scatter(y_test,y_predict)
```

```python
cost = MSE(y_predict,y_test)/(2*np.shape(y_predict)[0])
print(f"The cost is: {cost} and the median price is {np.median(Y_values)}")
print(f'This indicates a cost of {cost/np.median(Y_values*100)}%')
```
